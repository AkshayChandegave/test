<h3>Q2. Write a function to reverse a linked list in-place.</h3>
<p><b>Answer</b></p>

```python
## Linked list
# singly linked list.

class Node:
    
    def __init__(self, data=None, next=None):
        self.data = data 
        self.next = next
        
    def set_data(self, data):
        self.data = data
        
    def get_data(self):
        return self.data
    
    def set_next(self, next):
        self.next = next
        
    def get_next(self):
        return self.next          


## creating nodes.
head = Node(1, Node(2, Node(3, Node(4, Node(5,Node(6, Node(7)))))))


def traversal(head):
    current=head
    if current.data == None:
        print('Empty')
        
    while current:
        print(current.get_data(), end=' -> ')
        current = current.get_next()    
    return None
    
def reversed_in_place(head):
    
    prev=None
    current = head
    if current.data == None:
        return f'Empty'
    
    while current is not None:
        next_node = current.next
        current.next = prev
        
        prev = current
        current = next_node
    

    return traversal(head=prev)

print('Linked List before reversed:')
print(traversal(head=head))
print('\n')
print('Linked List after reversed in place:')
print(reversed_in_place(head=head))          

```

<h3>Q3. Detect cycle in linked list.</h3>
<p><b>Answer</b></p>
<p><em>A cycle in a linked list occurs when a node's next pointer points back to a previous node in the list, rather than None. This creates a loop where traversal of the list can continue indefinitely, never reaching the end. A linked list with a cycle does not have a None as the next reference in the last node.</em></p>

Example of a Cycle
<p>Consider a linked list with nodes A → B → C → D → E, where E.next points back to B instead of being None. <br>This creates a cycle: E → B → C → D → E...</p>


```python
## How to detect the cycle in linked list.

## define a linked list.

class linked_list:
    
    def __init__(self, data=None, next=None) -> None:
        self.data=data
        self.next= next
        
    def set_data(self, data):
        self.data=data
    
    def get_data(self):
        return self.data
    def set_next(self, next):
        self.next=next
    
    def get_next(self):
        return self.next
    

## creating nodes with cycle.

head=linked_list('A')
node2=linked_list('B')
node3=linked_list('C')
node4=linked_list('D')
node5=linked_list('E')

## connecting nodes
head.set_next(node2)
node2.set_next(node3)
node3.set_next(node4)
node4.set_next(node5)
node5.set_next(node2)  ## here E connect to B, it will create a cycle in linked list.

# By using Floyd's Cycle-Finding Algorithm is a popular method to detect cycles in a linked list.
# The idea is to use two pointers that traverse the list at different speeds:
# Slow Pointer (Tortoise): Moves one step at a time.
# Fast Pointer (Hare): Moves two steps at a time.

def detect_cycle(head):
    
    slow=head
    fast=head
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        
        if slow == fast:
            return f'Cycle found in this linked list.'
        
    return f'No cycle present in this linked list.'    
 
# Example.   
print(detect_cycle(head=head))
                
```

<h3>Q4. Merge two sorted linked list into one.</h3>
<p><b>Answer.</b></p>

```python
# Declaring class Node as linklist
class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next

    def set_data(self, data):
        self.data = data

    def get_data(self):
        return self.data

    def set_next(self, next):
        self.next = next

    def get_next(self):
        return self.next

def merge_sorted_lists(list1, list2):
    merged_head = Node()  # Dummy node to simplify the code
    current = merged_head

    while list1 is not None and list2 is not None:
        if list1.get_data() < list2.get_data():
            current.set_next(list1)
            list1 = list1.get_next()
        else:
            current.set_next(list2)
            list2 = list2.get_next()

        current = current.get_next()

    # Append remaining nodes from list1 or list2, if any
    if list1 is not None:
        current.set_next(list1)
    elif list2 is not None:
        current.set_next(list2)

    return merged_head.get_next()

# Example
list1_head = Node(1, Node(3, Node(5)))
list2_head = Node(2, Node(4, Node(6)))

merged_head = merge_sorted_lists(list1_head, list2_head)

print("Merged List:")
current = merged_head
while current is not None:
    print(current.get_data(), end=" -> ")
    current = current.get_next()
```

<h3>Q5. Write a function to remove nth node from the end in a linked list</h3>
<p><b>Answer.</b></p>

```python
# Declaring class Node as linklist
class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next

    def set_data(self, data):
        self.data = data

    def get_data(self):
        return self.data

    def set_next(self, next):
        self.next = next

    def get_next(self):
        return self.next

def remove_nth_from_end(head, n):
    dummy = Node()  # Create a dummy node to simplify the code
    dummy.set_next(head)
    fast = dummy
    slow = dummy

    # Move fast pointer n steps ahead
    for _ in range(n):
        fast = fast.get_next()

    # Move both pointers until fast reaches the end
    while fast.get_next() is not None:
        fast = fast.get_next()
        slow = slow.get_next()

    # Remove the nth node from the end
    slow.set_next(slow.get_next().get_next())

    return dummy.get_next()


linked_list_head = Node(1, Node(2, Node(3, Node(4, Node(5)))))

n = 1
new_head = remove_nth_from_end(linked_list_head, n)

print(f"Linked List after removing the {n}st node from the end:")
current = new_head
while current is not None:
    print(current.get_data(), end=" -> ")
    current = current.get_next()
```

<h3>Q6. Remove duplicates from a sorted linked list.</h3>
<p><b>Answer</b></p>

```python
# Define a linked list node
class linked_list:
    def __init__(self, data=None, next=None) -> None:
        self.data = data 
        self.next = next

    def set_data(self, data):
        self.data = data

    def get_data(self):
        return self.data

    def set_next(self, next):
        self.next = next

    def get_next(self):
        return self.next


def traversal(head):
    if head is None:  
        print('Empty list')
        return
    
    current = head
    while current:
        print(current.get_data(), end=' -> ')
        current = current.get_next()
    
# Function to remove duplicates from a sorted linked list
def remove_duplicate(head):
    current = head
    while current and current.get_next():  
        if current.get_data() == current.get_next().get_data():
            current.set_next(current.get_next().get_next())
        else:
            current = current.get_next()  
            
# Creating nodes with duplicate values (sorted linked list)
head = linked_list(1, linked_list(2, linked_list(2, linked_list(3, linked_list(3, linked_list(4, linked_list(5)))))))

# Example.
print("Original list:")
traversal(head)  

remove_duplicate(head)
print('\n')
print("List after removing duplicates:")
traversal(head)  
```

<h3>Q7. Find the intersection of two linked list.</h3>
<p><b>Answer.</b></p>

```python

# Declaring class Node as linklist
class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next

    def set_data(self, data):
        self.data = data

    def get_data(self):
        return self.data

    def set_next(self, next):
        self.next = next

    def get_next(self):
        return self.next

def find_intersection_point(head1, head2):
    def get_length(node):
        length = 0
        while node is not None:
            length += 1
            node = node.get_next()
        return length

    len1 = get_length(head1)
    len2 = get_length(head2)

    # Move the pointers to the same starting point
    current1, current2 = head1, head2
    while len1 > len2:
        current1 = current1.get_next()
        len1 -= 1
    while len2 > len1:
        current2 = current2.get_next()
        len2 -= 1

    # Traverse both lists until an intersection is found
    while current1 is not None and current2 is not None:
        if current1 == current2:
            return current1
        current1 = current1.get_next()
        current2 = current2.get_next()

    # If no intersection is found
    return None

# Example usage
common_node = Node(7, Node(8, Node(9)))

list1_head = Node(1, Node(2, common_node))
list2_head = Node(3, Node(4, Node(5, common_node)))

intersection_point = find_intersection_point(list1_head, list2_head)

if intersection_point:
    print(f"Intersection Point found at value {intersection_point.get_data()}")
else:
    print("No Intersection Point found.")
```

<h3>Q8. Rotate a linked list by k position to the right.</h3>
<p><b>Answer.</b></p>

```python
# Declaring class Node as linklist
class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next

    def set_data(self, data):
        self.data = data

    def get_data(self):
        return self.data

    def set_next(self, next):
        self.next = next

    def get_next(self):
        return self.next

def rotate_linked_list(head, k):
    if head is None or k == 0:
        return head

    # Find the length of the linked list
    length = 1
    current = head
    while current.get_next() is not None:
        length += 1
        current = current.get_next()

    # Determine the effective rotation amount
    k = k % length
    if k == 0:
        return head

    # Find the new tail and head after rotation
    new_tail = head
    for _ in range(length - k - 1):
        new_tail = new_tail.get_next()

    new_head = new_tail.get_next()
    new_tail.set_next(None)

    # Connect the original tail to the original head
    current.set_next(head)

    return new_head

# Example usage
linked_list_head = Node(1, Node(2, Node(3, Node(4, Node(5)))))
k_value = 2

rotated_head = rotate_linked_list(linked_list_head, k_value)

print(f"Linked List after rotating to the right by {k_value} places:")
current = rotated_head
while current is not None:
    print(current.get_data(), end=" -> ")
    current = current.get_next()
```

<h3>Q9. Add Two numbers represented by linked lists.</h3>
<p><b>Answer.</b></p>

```python
# Declaring class Node as linklist
class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next

    def set_data(self, data):
        self.data = data

    def get_data(self):
        return self.data

    def set_next(self, next):
        self.next = next

    def get_next(self):
        return self.next

def add_linked_lists(list1, list2):
    dummy = Node()  
    current = dummy
    carry = 0

    while list1 is not None or list2 is not None or carry:
        
        val1 = list1.get_data() if list1 is not None else 0
        val2 = list2.get_data() if list2 is not None else 0

  
        total = val1 + val2 + carry
        carry = total // 10

        # Create a new node with the sum
        current.set_next(Node(total % 10))
        current = current.get_next()

        # Move to the next nodes in both lists if available
        if list1 is not None:
            list1 = list1.get_next()
        if list2 is not None:
            list2 = list2.get_next()

    return dummy.get_next()

# Example usage
list1 = Node(2, Node(4, Node(3)))  
list2 = Node(5, Node(6, Node(4)))  

result = add_linked_lists(list1, list2)

print("Sum:")
current = result
while current is not None:
    print(current.get_data(), end="")
    current = current.get_next()

```

<h3>Q10. Clone a Linked List with next and Random pointer.</h3>
<p><em>Given a linked list of size N where each node has two links: one pointer points to the next node and the
second pointer points to any node in the list. The task is to create a clone of this linked list in O(N) time. </em></p>
<p><b>Answer.</b></p>

```python
class Node:
    def __init__(self, data, next=None, random=None):
        self.data = data
        self.next = next
        self.random = random

def clone_linked_list(head):
    if head is None:
        return None

    # Step 1: Insert cloned nodes just after original nodes
    current = head
    while current:
        new_node = Node(current.data)
        new_node.next = current.next
        current.next = new_node
        current = new_node.next

    # Step 2: Assign random pointers to the cloned nodes
    current = head
    while current:
        if current.random:
            current.next.random = current.random.next
        current = current.next.next

    # Step 3: Separate the cloned list from the original list
    current = head
    clone_head = head.next
    while current:
        clone_node = current.next
        current.next = clone_node.next
        if clone_node.next:
            clone_node.next = clone_node.next.next
        current = current.next

    return clone_head

def print_linked_list(head):
    current = head
    while current:
        random_data = current.random.data if current.random else 'None'
        print(f'Node data: {current.data}, Random points to: {random_data}')
        current = current.next

# Example usage:
# Creating a linked list with next and random pointers
original = Node(1)
original.next = Node(2)
original.next.next = Node(3)
original.next.next.next = Node(4)

# Setting up random pointers
original.random = original.next.next  # 1's random points to 3
original.next.random = original  # 2's random points to 1
original.next.next.random = original.next.next.next  # 3's random points to 4
original.next.next.next.random = original.next  # 4's random points to 2



# Cloning the linked list
cloned_list = clone_linked_list(original)

print("\nCloned list:")
print_linked_list(cloned_list)
```
